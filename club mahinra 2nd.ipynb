```python
train.columns
from datetime import datetime
stay_day=[]
for x in range(0,len(train.booking_date)):
    days=datetime.strptime(train.checkout_date[x],'%d/%m/%y')-datetime.strptime(train.checkin_date[x],'%d/%m/%y')
    stay_day.append(days.days)
train['stay_day']=stay_day
stay_year=[]
for x in range(0,len(train.booking_date)):
    year=datetime.strptime(train.booking_date[x],'%d/%m/%y')
    stay_year.append(year.year)
train['stay_year']=stay_year
```
